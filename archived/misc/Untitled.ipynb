```python
import json
```

```python
data = json.load(open("output.json", "r"))
```

```python
data = {tuple(json.loads(k)): set(tuple(i) for i in v) for k, v in data.items()}
```

```python
invariant = (0.9, 0.9, 0.9)
```

```python
goes_to_invariant = {i for i in data if invariant in data[i]}
```

```python
data_goes_to_invariant = {k: data[k] for k in goes_to_invariant}
data_goes_to_invariant
```
