```python
paths = json.load(open("output.json", "r"))
paths
```

```python
G = nx.Graph()
G.add_nodes_from(paths)
```

```python
def get_color_map(G):
    color_map = []
    for node in G:
        if node in invariants:
            color_map.append('green')
        else: 
            color_map.append('blue')
    return color_map
```

```python
def draw_graph(G):
    nx.draw(G, with_labels=True, node_color=get_color_map(G), font_color='white', font_weight='bold', node_size=3000)
```

```python
draw_graph(G)
```

```python
for k, v in paths.items():
    for iv in v:
        G.add_edge(k, tuple(iv)) 
```

```python
draw_graph(G)
```
