```python
import matplotlib.pyplot as plt
```

```python
model_name = "lstm_trained_at_2025_04_07_16_40"

graph_name = "graph_powerlaw_cluster_graph_n7"
```

```python
pred_file_name = f"{model_name}__rank_effect_ml_pred__{graph_name}"
```

```python
# Read and parse the data
with open(f"{pred_file_name}.txt", "r") as file:
    data = file.read()
    numbers = list(
        map(float, data.strip(",").split(","))
    )  # Convert to float or int as needed

# Plot histogram
plt.hist(numbers, edgecolor="black")
plt.title("Histogram of Data")
plt.xlabel("Rank Effect")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
