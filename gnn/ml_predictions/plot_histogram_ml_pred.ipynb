```python
import matplotlib.pyplot as plt
```

```python
model_name = "lstm_trained_at_2025_04_10_00_11"

graph_name = "star_graph_n15"
```

```python
pred_file_name = f"{model_name}__rank_effect_ml_pred__{graph_name}"
```

```python
# Read and parse the data
with open(f"{pred_file_name}.txt", "r") as file:
    data = file.read()
    numbers = list(
        map(float, data.strip(",").split(","))
    )  # Convert to float or int as needed

# Plot histogram
plt.hist(numbers, edgecolor="black")
plt.title(f"Rank Effect v/s Count - {graph_name}")
plt.xlabel("Rank Effect")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```
