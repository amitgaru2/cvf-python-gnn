```python
import torch

import pandas as pd
```

```python
df = pd.read_csv("datasets/custom_data__m2_35__c10_5.csv")
# df = pd.read_csv("datasets/SOCR-HeightWeight-XY.csv")
```

```python
df
```

```python
X = torch.tensor(df["X"])
y = torch.tensor(df['y'])
X, y
```

```python
lr = 0.001

m = torch.tensor(3.0, requires_grad=True)
c = torch.tensor(0.0, requires_grad=True)

for i in range(10000):
    y_pred = m * X + c
    loss = ((y_pred - y) ** 2).mean()
    loss.backward()


    # m = m - lr * m.grad
    # c = c - lr * c.grad
    with torch.no_grad():
        m -= lr * m.grad
        c -= lr * c.grad
        m.grad.zero_()
        c.grad.zero_()
    
    
    # print(m, c)
```
